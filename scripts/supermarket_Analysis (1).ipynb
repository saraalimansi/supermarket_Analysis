```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df=pd.read_csv('/content/supermarket_sales .csv')
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
df
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
new_df=df.copy()
new_df.head(10)
```

```python
plt.figure(figsize=(15, 6))
sns.boxplot(new_df)
plt.title("Box Plot of Rating")
plt.xlabel("Rating")
plt.ylabel("Count")
plt.show()
```

```python
new_df.tail(10)
```

```python
#Delete fixed values
new_df = new_df.loc[:, new_df.nunique() > 1]
```

```python
new_df[['Tax 5%','Total','gross income']]=new_df[['Tax 5%','Total','gross income']].round(2)
new_df.head()
```

```python
# Extract numeric columns only
numeric_df = new_df.select_dtypes(include=['number'])

plt.figure(figsize=(10, 6))
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

```python
#Distribution of Ratings
plt.figure(figsize=(8, 5))
sns.histplot(new_df['Rating'], bins=15, kde=True, color='blue')
plt.title("Distribution of Customer Ratings")
plt.xlabel("Rating")
plt.ylabel("Frequency")
plt.show()
```

```python
#Payment Method Distribution
plt.figure(figsize=(8, 5))
sns.countplot(x='Payment', data=new_df, palette="pastel")
plt.title("Distribution of Payment Methods")
plt.xlabel("Payment Method")
plt.ylabel("Count")
plt.show()
```

```python
#Sales by Product Line
plt.figure(figsize=(10, 6))
sns.barplot(x=new_df['Product line'], y=new_df['Total'], ci=None, estimator=sum, palette="viridis")
plt.title("Total Sales by Product Line")
plt.xlabel("Product Line")
plt.ylabel("Total Sales")
plt.xticks(rotation=45)
plt.show()
```

```python
#new_df['Date'].dtype
new_df['Date'] = pd.to_datetime(new_df['Date'])
new_df['Month'] = new_df['Date'].dt.month.astype(str)
new_df.head()
```

```python
#trends of sales
plt.figure(figsize=(10, 6))
sns.lineplot(x='Month', y='Total', data=new_df, marker='o', color='blue')
plt.title("trends of sales ber months")
plt.xlabel("Month")
plt.ylabel("Total Sales")
plt.xticks(rotation=45)
plt.show()
```

```python
new_df['Hour']=pd.to_datetime(new_df['Time']).dt.hour

new_df['Hour_12'] = new_df['Hour'] % 12
new_df['Hour_12'] = new_df['Hour_12'].replace(0, 12)

new_df['Hour_12'].head()
```

```python
#Sales Trends by Hour
plt.figure(figsize=(10, 6))
sns.lineplot(x=new_df['Hour'], y=new_df['Total'], estimator='sum', ci=None, color='green')
plt.title("Sales Trends by Hour")
plt.xlabel("Hour of the Day")
plt.ylabel("Total Sales")
plt.grid(True)
plt.show()
```

```python
#Distribution of Gender
plt.figure(figsize=(10, 6))
plt.pie(new_df['Gender'].value_counts(), labels=new_df['Gender'].value_counts().index, autopct='%1.1f%%', startangle=90, colors=['lightblue', 'lightcoral'])
plt.axis('equal')
plt.title("Distribution of Gender")
plt.show()
```

```python
#Distribution of Customer Types
plt.figure(figsize=(10, 6))
sns.countplot(x='Customer type', data=new_df, palette="pastel")
plt.title("Distribution of Customer Types")
plt.xlabel("Customer Type")
plt.ylabel("Count")
plt.show()

```

```python
#Distribution of Cities
plt.figure(figsize=(10, 6))
sns.countplot(x='City', data=new_df, palette="pastel")
plt.title("Distribution of Cities")
plt.xlabel("City")
plt.ylabel("Count")
plt.show()
```

```python
# Distribution of Branch with Total Sales
plt.figure(figsize=(8, 5))
sns.barplot(x='Branch', y='Total', data=new_df, estimator=sum, ci=None, palette="Set3")
plt.title("Branch-Wise Total Sales")
plt.xlabel("Branch")
plt.ylabel("Total Sales")
plt.show()
```
